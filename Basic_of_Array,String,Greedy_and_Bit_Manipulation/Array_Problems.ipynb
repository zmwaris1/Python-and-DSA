```python
# Find a peak element which is not smaller than its neighbours

def findPeak(arr, n):
    
    if n == 1:
        return 0
    if (arr[0]>=arr[1]):
        return 0
    if arr[n-1] >= arr[n-2]:
        return n - 1
    
    for i in range(1, n-1):
        if (arr[i] >= arr[i-1] and arr[i] >= arr[i+1]):
            return i

    return -1

def findPeakRecurse(arr, low, high, n):

    mid = low + (high - low) // 2

    if ((mid == 0 or arr[mid - 1] <= arr[mid]) and (mid == n-1
                or arr[mid+1] <= arr[mid])):
        return mid
    elif (mid > 0 and arr[mid-1] > arr[mid]):
        return findPeakRecurse(arr, low, mid-1, n)
    else:
        return findPeakRecurse(arr, mid+1, high, n)

def findPeakIterative(arr, n):

    l = 0
    r = n-1

    while (l <= r):
        mid = (l + r) >> 1

        if ((mid == 0 or arr[mid - 1] <= arr[mid]) and (mid == n-1 or arr[mid+1]<=arr[mid])):
            break

        if (mid > 0 and arr[mid-1] > arr[mid]):
            r = mid - 1
        else:
            l = mid + 1
    return mid
```

```python
# Find the minimum (or maximum) element of an array

class MinMax:

    def getMin(self, arr, n):
        res = arr[0]
        for i in range(1, n):
            res = min(res, arr[i])
        return res

    def getMax(self, arr, n):
        res = arr[0]
        for i in range(1, n):
            res = max(res, arr[i])
        return res
```

```python
def reverseList(A, start, end):
    while start < end:
        A[start], A[end] = A[end], A[start]
        start += 1
        end -= 1
    return A

def reverseListRe(A, start, end):
    if start >= end:
        return
    A[start], A[end] = A[end], A[start]
    reverseListRe(A, start+1, end-1)

def reverseListUsingStack(arr):
    stack = []

    for ele in arr:
        stack.append(ele)
    
    for i in range(len(arr)):
        arr[i] = stack.pop()
    
    return arr
```

```python
# Sort a given array

def selectionSort(arr, n):
    
    for i in range(n):
        min_idx = i

        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j 

        arr[min_idx], arr[i] = arr[i], arr[min_idx]
    return arr
```

```python
def countOccurences(arr, n, x):
    res = 0
    for i in range(n):
        if x == arr[i]:
            res += 1
    return res

def firstOcc(arr, l, h, x):
    if h >= l:
        mid = l + (h - l) // 2

        if (mid == 0 or x > arr[mid - 1]) and arr[mid] == x:
            return mid
        elif x > arr[mid]:
            return firstOcc(arr, mid+1, h, x)
        else:
            return firstOcc(arr, l, mid-1, x)
    return -1

def lastOcc(arr, n, l, h, x):
    if h >= l:
        mid = l + (h-l)//2

        if (mid == n-1 or x < arr[mid+1]) and arr[mid] == x:
            return mid
        elif x < arr[mid]:
            return lastOcc(arr, n, l, mid-1, x)
        else:
            return lastOcc(arr, n, mid+1, h, x)
    return -1

def countOccurences(arr, n, x):
    idxFirst = firstOcc(arr, 0, n-1, x)
    if idxFirst == -1:
        return -1
    idxLast = lastOcc(arr, n, idxFirst, n-1, x)
    return idxLast - idxFirst + 1
```

```python
# Sort an array of 0s, 1s and 2s. Dutch National Flag Problem

def sort012(arr):
    lo = 0
    hi = len(arr) - 1
    mid = 0

    while mid <= hi:
        if arr[mid] == 0:
            arr[lo], arr[mid] = arr[mid], arr[lo]
            lo += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[mid], arr[hi] = arr[hi], arr[mid]
            hi += 1
    return arr
```

```python
# Find subarray with given sum

def subArraySum(arr, n, sum):
    for i in range(0, n):
        curSum = arr[i]
        if curSum == sum:
            return
        else:
            for j in range(i+1, n):
                curSum += arr[j]
                if curSum == sum:
                    return
    return -1

def subArraySum(arr, n, sum):
    curSum = arr[0]
    start = 0

    i = 1
    while i <= n:
        while curSum > sum and start < i - 1:
            curSum -= arr[start]
            start += 1
        if curSum == sum:
            return 1
        if i < n:
            curSum += arr[i]
        i += 1
    return 0
```

```python
# Move all negative numbers to beginning and positive to end with constant extra space

# Two-Pointer Approach

def shift(arr, left, right):
    while left <= right:
        if arr[left] < 0 and arr[right] < 0:
            left += 1
        elif arr[left] > 0 and arr[right] < 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right += 1
        elif arr[left] > 0 and arr[right] > 0:
            right -= 1
        else:
            left += 1
            right -= 1
    
    return arr

# Dutch National Flag Approach

def reArrange(arr, n):
    low, high = 0, n-1
    while (low < high):
        if arr[low] < 0:
            low += 1
        elif arr[high] > 0:
            high -= 1
        else:
            arr[low], arr[high] = arr[high], arr[low]
    
    return arr
```

```python
# Program to cyclically rotate an array by one

def rotateEnd(arr, n):
    last = arr[n-1]

    for i in range(n-1, 0, -1):
        arr[i] = arr[i-1]
    
    arr[0] = last

    return arr

def rotateEnd(arr, n):
    i = 0
    j = n-1

    while i != j:
        arr[i], arr[j] = arr[j], arr[i]
        i += 1
    return arr
```

```python
# Find the Missing Number

def findMissing(arr, N):
    temp = [0] * (N+1)

    for i in range(0, N):
        temp[arr[i] - 1] = 1
    for i in range(0, N+1):
        if temp[i] == 0:
            ans = i + 1
    return ans
```

```python
# Count pairs with given sum

def getPairsCount(arr, n, k):
    count = 0

    for i in range(0, n):
        for j in range(i+1, n):
            if arr[i] + arr[j] == k:
                count += 1
    return count

# Using Binary-Search

class CountPairs:

    def lowerBound(self, arr, start, end, key):
        while start < end:
            mid = start + (end - start) // 2
            if arr[mid] < key:
                start = mid + 1
            else:
                end = mid
        
        return start
    
    def upperBound(self, arr, start, end, key):
        while start < end:
            mid = start + (end - start) // 2
            if arr[mid] <= key:
                start = mid + 1
            else:
                end = mid
        
        return start
    
    def getPairsCount(self, arr, n, k):
        arr = arr.sort()
        c = 0
        for i in range(0, n-1):
            x = k - arr[i]
            y = self.lowerBound(arr, i+1, n, x)
            z = self.upperBound(arr, i+1, n, x)
            c = c + z - y
        
        return c
    
# Using Hashing

def getPairsCount(arr, n, sum):
    umap = {}
    count = 0
    for i in range(n):
        if sum - arr[i] in umap:
            count += umap[sum-arr[i]]
        if arr[i] in umap:
            umap[arr[i]] += 1
        else:
            umap[arr[i]] = 1
    return count
```

```python
# Find duplicates in O(n) time and O(1) extra space

def duplicates(arr):
    umap = {}
    result = []

    for num in arr:
        umap[num] = umap.get(num, 0) + 1
    
    for key, value in umap.items():
        if value > 1:
            result.append(key)
    
    if not result:
        result.append(-1)
    result.sort()
    return result
```

```python
# Quicksort

def partition(arr, low, high):
    pivot = arr[high]
    i = low -1

    for j in range(low, high):
        if arr[i] <= pivot:
            i += 1
            arr[i], arr[j], arr[j], arr[i]

    arr[i+1], arr[high] = arr[high], arr[i+1]

    return i+1

def quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)

        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
```

```python
# Find common elements in three sorted arrays
# Naive Approach
def commonElements(A, B, C):
    count = {}

    for ele in A:
        count[ele] = 1
    
    for ele in B:
        if ele in count and count[ele] == 1:
            count[ele] = 2
    
    for ele in C:
        if ele in count and count[ele] == 2:
            count[ele] = 3
    
    comm = []
    for ele, cnt in sorted(count.items()):
        if cnt == 3:
            comm.append(ele)
    
    return comm

# Expected Approach
def commonElements(A, B, C):
    i, j, k = 0, 0, 0
    comm = []

    while i < len(A) and j < len(B) and k < len(C):
        if A[i] == B[j] and B[j] == C[k]:
            comm.append(A[i])
            i += 1
            j += 1
            k += 1

            while i < len(A) and A[i] == A[i-1]:
                i += 1
            while j < len(B) and B[j] == B[j-1]:
                j += 1
            while k < len(C) and C[k] == C[k-1]:
                k += 1
        elif A[i] < B[j]:
            i += 1
        elif B[j] < C[k]:
            j += 1
        else:
            k += 1
    return comm
```

```python
# Find the first repeating element in an array of integers

# Using Nested Loop
def firstRepeatingElement(arr, n):
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] == arr[j]:
                return arr[i]
    return -1

# Using Hashset
def firstRepeatingElement(arr, n):
    mini = -1
    myset = dict()

    for i in range(n-1, -1, -1):
        if arr[i] in myset.keys():
            mini = i
        else:
            myset[arr[i]] = 1
    
    if mini != -1:
        return arr[mini]
    else:
        return None
```

```python
# Find first non-repeating element in a given Array of integers

# Using Hashing
from collections import defaultdict

def firstNonRepeating(arr, n):
    mp = defaultdict(lambda: 0)

    for i in range(n):
        mp[arr[i]] += 1

    for i in range(n):
        if mp[arr[i]] == 1:
            return arr[i]
    return -1
```

```python
def countSubArrWithEqualZeroAndOne(arr, n):
    mp = dict()
    Sum = 0
    count = 0

    for i in range(n):
        if (arr[i] == 0):
            arr[i] = -1
        Sum += arr[i]

        if (Sum == 0):
            count += 1
        
        if (Sum in mp.keys()):
            count += mp[Sum]
        
        mp[Sum] = mp.get(Sum, 0) + 1
    
    return count
```

```python
# Rearrange array in alternating positive & negative items

def rightRotate(arr, n, outOfPlace, cur):
    temp = arr[cur]
    for i in range(cur, outOfPlace, -1):
        arr[i] = arr[i-1]
    arr[outOfPlace] = temp
    return arr

def rearrange(arr, n):
    outOfPlace = -1
    for index in range(n):
        if outOfPlace >= 0:
            if ((arr[index] >= 0 and arr[outOfPlace] < 0) or
                (arr[index] < 0 and arr[outOfPlace] >= 0)):
                arr = rightRotate(arr, n, outOfPlace, index)
                if index - outOfPlace > 2:
                    outOfPlace += 2
                else:
                    outOfPlace = -1
        if outOfPlace == -1:
            if ((arr[index] >= 0 and index % 2 == 0) or
                (arr[index] < 0 and index % 2 == 1)):
                outOfPlace = index
    return arr
```
