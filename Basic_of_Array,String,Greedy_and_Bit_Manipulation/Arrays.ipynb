### Reverse An Array

```python
# 1. Reverse using an extra array

def reversed_array(arr):
    reversed_arr = arr[::-1]
    return reversed_arr

# 2. Using a loop

def reverseList(A, start, end):
    while start < end:
        A[start], A[end] = A[end], A[start]
        start += 1
        end -= 1
    return A

# 3. Reverse using in-built methods

def reverse(arr):
    reverse_arr = list(reversed(arr))
    return reverse_arr

# 4. Reverse array using recursion

def reverseList(A, start, end):
    if start >= end:
        return
    A[start], A[end] = A[end], A[start]
    reverseList(A, start+1, end-1)

# 5. Reverse array using Stack

def reverse_arr_stack(arr):
    stack = []
    for ele in arr:
        stack.append(ele)
    for i in range(len(arr)):
        arr[i] = stack.pop()

    return arr
```

### Sum Of Digits

```python
# 1. Sum of digits using remainder method

def getSum(n):
    sum = 0

    while n > 0:
        sum += n%10
        n = n//10

    return sum

def getSumRecur(n):
    if n == 0:
        return 0
    
    sum = n%10 + getSumRecur(n//10)
    return sum
```

### Check if given string is Palindrome

```python
def isPalindrome(s):
    rev = s[::-1]
    if s == rev:
        return True
    return False

def isPalindromeLoop(s):
    for i in range(len(s//2)):
        if str[i] != str[len(s)-i-1]:
            return False
    return True
```

### Sum of elements in a given array

```python
# 1. Sum of array elements using recursion

def sumEle(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        return arr[0] + sumEle(arr[1:])

# 2. Sum of array elements using Iteration

def sumEleI(arr):
    sum = 0
    for i in range(len(arr)):
        sum += arr[i]

    return sum

# 3. Sum of array elements using Divide and Conquer

def sumEleDandC(arr, low, high):
    if low == high:
        return arr[low]
    else:
        mid = (high + low)//2
        left_sum = sumEleDandC(arr, low, mid)
        right_sum = sumEleDandC(arr, mid+1, high)

        return left_sum + right_sum
```

```python
# Maximum and minimum in an array

# 1. Using functions

def set_min(A):
    mini = float('inf')
    for num in A:
        if num < mini:
            mini = num
    return mini

def set_max(A):
    maxi = float('-inf')
    for num in A:
        if num > maxi:
            maxi = num
    return maxi

# 2. Using Linear Search

class Pair:
    def __init__(self):
        self.min = 0
        self.max = 0

def getMinMax(arr, n):
    minmax = Pair()

    if n == 1:
        minmax.max = arr[0]
        minmax.min = arr[0]
        return minmax
    
    if arr[0] > arr[1]:
        minmax.min = arr[1]
        minmax.max = arr[0]
    else:
        minmax.min = arr[0]
        minmax.min = arr[1]

    for i in range(2, n):
        if arr[i] > minmax.max:
            minmax.max = arr[i]
        elif arr[i] < minmax.min:
            minmax.min = arr[i]
    return minmax

# 3. Using Tournament method

def getMinMax(low, high, arr):
    arr_max = arr[low]
    arr_min = arr[low]

    if low == high:
        return (arr_max, arr_min)
    elif high == low+1:
        if arr[low] > arr[high]:
            arr_max = arr[low]
            arr_min = arr[high]
        else:
            arr_max = arr[high]
            arr_min = arr[low]
        return (arr_max, arr_min)
    
    mid = (low + high) // 2
    arr_max1, arr_min1 = getMinMax(low, mid, arr)
    arr_max2, arr_min2 = getMinMax(mid+1, high, arr)

    return (max(arr_max1, arr_max2), min(arr_min1, arr_min2))
```

```python
# Prefix Sum

def prefixSum(arr: list, n: int, prefixSum: list) -> list:
    prefixSum[0] = arr[0]

    for i in range(1, n):
        prefixSum[i] = prefixSum[i-1] + arr[i]
    
    return prefixSum

# Equilibrium index of an array - An index such that the sum of elements at lower indexes is equal to the sum of elements at higher indexes

def equilibrium(arr: list) -> int:
    n = len(arr)

    for i in range(n):
        leftSum = 0
        rightSum = 0
        for j in range(i):
            leftSum += arr[i]

        for j in range(i+1, n):
            rightSum += arr[j]

        if leftSum == rightSum:
            return i
    
    return -1

# Equilibrium index using Array-Sum

def equilibrium(arr):

    total_sum = sum(arr)

    leftSum = 0
    for i, num in enumerate(arr):
        total_sum -= num

        if leftSum == total_sum:
            return i
        leftSum += num
    return -1

# Equilibrium index using Prefix Sum

def equilibrium(arr):
    left_sum = [arr[0]]
    right_sum = [arr[len(arr)-1]]

    for i in range(1, len(arr)):
        left_sum.append(left_sum[i-1]+arr[i])
        right_sum.append(right_sum[i-1]+arr[len(arr)-1-i])

    for i in range(len(arr)):
        if left_sum[i] == right_sum[len(arr)-i-1]:
            return i
        
    return -1

# Eq index using two pointers

def pivotIndex(nums: list[int]) -> int:
    left, pivot, right = 0, 0, sum(nums)-nums[0]
    while pivot < len(nums) - 1 and right != left:
        pivot += 1
        right -= nums[pivot]
        left += nums[pivot-1]

    return pivot if left == right else -1
```

```python
# 2D - Prefix Sum

def prefixSum2D(a: list[list], C: int, R: int) -> list[list]:
    psa = [[0 for x in range(C)] for y in range(R)]
    psa[0][0] = a[0][0]

    for i in range(1, C):
        psa[0][i] = psa[0][i-1] + a[0][i]
    for i in range(0, R):
        psa[i][0] = psa[i-1][0] + a[i][0]

    for i in range(1, R):
        for j in range(1, C):
            psa[i][j] = psa[i-1]+psa[i][j-1]-psa[i-1][j-1]+a[i][j]
    
    return psa
```

```python
# Coordinate Compression - Convert an Array to reduced form using Hashing

"""Convert the given array to a form where all elements are in the range
from 0 to N-1. Order of the elements is same i.e 0 placed in place of smallest element and so on."""

import sys

def convert(arr: list[int]) -> list[int]:
    n = len(arr)
    result = [0] * n
    curr_pos = 0

    for i in range(n):
        minn = sys.maxsize
        idx = -1

        for j in range(n):
            if minn > arr[j]:
                minn = arr[j]
                idx = j

        result[idx] = curr_pos
        curr_pos += 1

        arr[idx] = sys.maxsize

    return result

# Efficient Approach

def convert(arr: list[int], n: int) -> list[int]:
    temp = [arr[i] for i in range(n)]
    temp.sort()

    umap = {}

    val = 0
    for i in range(n):
        umap[temp[i]] = val
        val += 1

    for i in range(n):
        arr[i] = umap[arr[i]]

    return arr
```

```python
# Kadane Algorithm - Largest Sum Contiguous array

def kadane(a: list[int], size: int) -> int:
    max_so_far = float('-inf')
    max_ending_here = 0

    for i in range(size):
        max_ending_here = max_ending_here + a[i]
        if max_so_far < max_ending_here:
            max_so_far = max_ending_here

        if max_ending_here < 0:
            max_ending_here = 0
    return max_so_far
```
